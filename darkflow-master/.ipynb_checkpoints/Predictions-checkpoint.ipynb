```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

```python
options = {
    'model': 'cfg/tiny-yolo-voc-1c.cfg',
    'load': 30,
    'threshold': 0.6,
    'labels': 'labels_thyroid.txt',
}

```

```python
tfnet = TFNet(options)
```

```python
img = cv2.imread('550-3.jpg', cv2.IMREAD_COLOR)

result = tfnet.return_predict(img)
result[0]['topleft']
result[0]
res = result[0]
box = res['topleft']['x'], res['topleft']['y'],res['bottomright']['x'],res['bottomright']['y']

box
res
```

```python
box_num = 0
t1 = (result[box_num]['topleft']['x'], result[box_num]['topleft']['y'])
br = (result[box_num]['bottomright']['x'], result[box_num]['bottomright']['y'])
label = result[box_num]['label']

img = cv2.rectangle(img, t1, br, (0,255,0), 8)
img = cv2.putText(img, label, t1, cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 2)



plt.imshow(img)
plt.show()
```
